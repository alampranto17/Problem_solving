### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# HISTPLOT
#FIGURE LEVEL --> DISPLOT
#AXES LEVEL --> HISTPLOT ,KDEPLOT, RUGPLOT
```

```python
# PLOTTING univeriate histogram ->(amra ekta column er upore focus korbo)

sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(kind='hist',data=tips,x='total_bill')
```

```python
sns.displot(kind='hist',data=tips,x='tip',hue='sex',element='step')
```

```python
titanic=sns.load_dataset('titanic')
titanic['survived']=titanic['survived'].apply(lambda x: 'Yes' if x==1 else 'No')
titanic
sns.displot(kind='hist',data=titanic,x='age',col='survived',element='step')
```

# kde plot

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(kind='hist',data=tips,x='total_bill',kde=True)
```

```python
sns.displot(kind='kde',data=tips,x='total_bill',hue='sex',fill=True)
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# heatmap
data_country=gaps.pivot_table(index='country',columns='year',values='lifeExp')
data_country

# axes level function kuno figure level function nai 
plt.figure(figsize=(15,15))
sns.heatmap(data_country)
```

```python
# Let's filter it i need only europe
europe=gaps[gaps['continent']=='Europe']

europe=europe.pivot_table(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(europe,annot=True,linewidths=0.5,cmap='viridis')
```
