```python
import re
def convert_to_minutes(text):
    # Look for hours
    hrs_match = re.search(r"(\d+)\s*(?:hr|hrs|hour|hours)", text)
    hours = int(hrs_match.group(1)) if hrs_match else 0

    # Look for minutes
    mins_match = re.search(r"(\d+)\s*(?:min|mins|minute|minutes)", text)
    minutes = int(mins_match.group(1)) if mins_match else 0

    return hours * 60 + minutes

df['time']=df.time.apply(convert_to_minutes)
```

```python
df
```

```python
df['rating']=df.stars.str.split('stars').str.get(0)
df['rating']=df['rating'].str.split(' ').str.get(0)
df['votes']=df['stars'].str.split('stars').str.get(1).str.split(' ').str.get(0)
```

```python
df['rating'].replace('Not',0,inplace=True)
df.info()
```

```python
df['rating']=df.rating.astype(np.float16)
df.info()
```

```python
df['votes'].replace('NaN',0,inplace=True)
```

```python

df['votes']=df['votes'].str.replace(',','').astype(np.float16)
```

```python
df['votes']=df['votes'].fillna(0).astype(np.int16)
```

```python
df.drop(columns='stars',inplace=True)
```
